```python
import pandas as pd

from matplotlib import pyplot as plt

data = pd.read_csv('data/lang_salaries.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# to fill from median instead of plotting from 0
overall_median = 57287

# alpha --> to set how much we can see through
# interpolate=True => will make sure that certain x regions are not going to be clipped and that all 
# are going to be filled. Defualt fill color
# plt.fill_between(ages, py_salaries, overall_median, 
#                  where=(py_salaries > overall_median), # filling only where the py plot is > median salary
#                  interpolate=True, alpha=0.25)


# # to fill < median salary with another color
# # color for custom color
# plt.fill_between(ages, py_salaries, overall_median, 
#                  where=(py_salaries <= overall_median), # filling only where the py plot is > median salary
#                  interpolate=True, color='red', alpha=0.25)

## to fill between 2 plots
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), # filling only where the py plot is > median salary
                 interpolate=True, alpha=0.25, label='Above Avg')


# to fill < median salary with another color
# color for custom color
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), # filling only where the py plot is > median salary
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')


plt.legend()

plt.title('Median Salary(USD) by Age')

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
