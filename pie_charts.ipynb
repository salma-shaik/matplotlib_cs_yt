#### Best to use with upto 5 categories. >5 better to use bar charts

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


## values don't have to add upto 100. Coz pie would get proportions of each out of whole no matter what
slices = [120, 80, 30, 20]
sl_labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
sl_colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f

plt.pie(slices, labels=sl_labels, colors=sl_colors, wedgeprops = {'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")

# may be works better for traditional vertical charts
plt.tight_layout
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels, wedgeprops = {'edgecolor': 'black'})

plt.show()
```

```python
#### Explode ####

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# how much we want to emphasize a particular slice
# 0 - will keep as is; any num beside 0 - will represent fraction of the radius that we will like to explode
sl_explode = [0, 0, 0, 0.1, 0] # explodes python slice. 0.5 - explodes by half of the radius


## startangle=90 => 1st value starts at 90 degree angle
## autopct = '%1.1f%%' => format str to specify to calculate %s
plt.pie(slices, labels=labels, explode=sl_explode, shadow=True, startangle=90, 
        autopct = '%1.1f%%', wedgeprops = {'edgecolor': 'black'})

plt.show()
```
